```python
import pandas as pd
import requests
```

```python
data = pd.read_csv("player_stats.csv")
data.head()
```

```python
from bs4 import BeautifulSoup
import os
```

```python
def get_file_path(character_name, folder_name="thumbnails"):
    cleaned_name = character_name.replace(" ", "_")
    file_path = f'{folder_name}/{cleaned_name}.jpg'
    return file_path

def get_img_url(character_page):
    page = requests.get(character_page)
    soup = BeautifulSoup(page.text, 'html.parser')
    img_block = soup.find("img", class_="pi-image-thumbnail")
    if img_block is not None:
        return img_block["src"]
    return None
    
def download_img(img_url, file_path):
    img_data = requests.get(img_url).content
    with open(file_path, 'wb+') as handler:
        handler.write(img_data)
        print(f"Saved to {file_path}")
```

```python
folder = "thumbnails"
if not os.path.exists(folder):
    os.mkdir(folder)

img_df_data = []
base_url = "https://haikyuu.fandom.com/wiki/"
for player in data.name:
    cleaned_name = player.replace(" ", "_")
    filename = cleaned_name + '.jpg'

    img_filepath = os.path.join(folder, filename)
    img_url = None
    if os.path.exists(img_filepath):
        print(f"File already exists ({img_filepath})")
    else:
        query_url = base_url + player.replace(" ", "_")
        img_url = get_img_url(query_url)
        if img_url is not None:
            download_img(img_url, img_filepath)

    # save img url and filename
    img_df_data.append([player, filename, img_url])
```

```python
img_df = pd.DataFrame(img_df_data, columns=headers)
img_df.head()
```

```python
img_df.to_csv("player_thumbnails.csv", index=False)
```
